```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# rmse 함수
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

```python
# 다양한 모델 비교를 위해 get scores 함수 생성
def get_scores(models, train, y):
    df = {}

    for model in models:
        # 모델 이름 획득
        model_name = model.__class__.__name__

        # train, test 데이터셋 분리
        # random_state를 사용하여 고정하고 train과 test 셋의 비율은 8:2로 합니다.
        x_train,x_test,y_train,y_test = train_test_split(train,y, test_size = 0.2 , random_state = 42)

        # 모델 학습
        model.fit(x_train,y_train)
        
        # 예측
        y_pred = model.predict(x_test)

        # 예측 결과의 rmse값 저장
        df[model_name] = rmse(y_test,y_pred)
        
        # data frame에 저장
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
    return score_df
```

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
# random seed
random_state=64

gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)

models = [gboost, xgboost, lightgbm, rdforest]
get_scores(models,train,y)
```

- 결과적으로 RMSE가 가장 낮은 LGBMRegressor의 하이퍼 파라미터를 조정하는 방향으로 진행하도록 한다

```python
#  grid search 함수
from sklearn.model_selection import GridSearchCV
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=-1):
    grid_model = GridSearchCV(model, param_grid=param_grid, \
                        scoring='neg_mean_squared_error', \
                        cv=5, verbose=verbose, n_jobs=n_jobs)

    grid_model.fit(train, y)
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    results = pd.DataFrame(params)
    results['score'] = score
    results['RMSE'] = np.sqrt(-1 * results['score'])
    results = results.rename(columns={'RMSE': 'RMSLE'})
    
    return results.sort_values("RMSLE")
    
```

```python
# 하이퍼 파라미터 조정
param_grid = {
    "num_leaves":[20,25,30],
    "learning_rate" : [0.1],
    'n_estimators': [350,400,500],
    'max_depth': [8,12,16],
    "boosting_type":['gbdt'],
    "feature_fraction":[0.5],
}

model = LGBMRegressor(random_state=64)
my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5)
```

- num_leaves 10, 20 ,25, 30 ,35 모두 진행한 결과 20,25에서 높게 나타남
- learning_rate : 0.1
- n_estimators ,100,200,250,300,350,400,500,1000, 모두 진행 결과 250-400을 왔다 갔다 함
- max_depth 4,8,12,16,20 모두 해본 결과 8,12,16 사이를 왔다 갔다 한다 
- boosting_type 'gbdt'가 random forest보다 결과가 좋음
- feature_fraction: 0.5,0.8 진행해본 결과 0.5가 가장 좋음

**위의 파라미터는 한번에 돌리면 결과를 확인하는데 오래걸려서 여러번 나눠서 실험하면서 진행했습니다.**
